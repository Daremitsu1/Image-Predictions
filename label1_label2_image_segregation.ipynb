```python
from tensorflow.keras.models import load_model

prediction_folder = 'WorkOrderDump/WorkOrder04082023'
model_saved = load_model('vgg16_pp_image_segregation_label_1_classification_model4_04_08_2023.h5')
destpath =  'DistributedFolder04082023'
os.mkdir(destpath + '/' + 'Lawn Maintenance')
os.mkdir(destpath + '/' + 'Door Task')
# # Function to process image for initial image segregation
classes = {0: 'Door Tasks', 1: 'Lawn Maintenance'}
img_process = 0
count  = 0
for item in os.listdir(prediction_folder):
    X = cv2.imread(prediction_folder + '/' + item)
    print("X.shape :  ",X.shape)
    X = cv2.resize(X,(210,210))
    print("X.shape : ",X.shape)
    X  = X.reshape((1,210,210,3))
    P = model_saved.predict(X)
    # predicted_class = np.argmax(P)
    # print("P : ",P)
    # print("predicted_class : ",predicted_class)
    P = np.argmax(P, axis=1)
    print("P : ",P)
    print("predicted_label:",P[0])
    
    
    if P[0] == 1:
        count = count + 1
        shutil.copy2(prediction_folder + '/' + item, destpath + '/'+'Lawn Maintenance')
    elif P[0] == 0:   
        shutil.copy2(prediction_folder + '/' + item, destpath + '/' +'Door Task')
    img_process = img_process + 1
    print("img_process : ", img_process)        
    print("count : ", count)
print("final_count : ",count)    
```
